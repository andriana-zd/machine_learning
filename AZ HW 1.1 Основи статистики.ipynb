**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
threshold = mean_of_purchases + 2 * std_purchase_sample

filtered_data = [(sum_purchases[i], visits[i]) for i in range(len(sum_purchases)) if sum_purchases[i] <= threshold]

filtered_purchases, filtered_visits = zip(*filtered_data)
print(filtered_purchases)
print(filtered_visits)
```

```python
mean_of_filtered_purchases = np.mean(filtered_purchases)
print(mean_of_filtered_purchases)
std_filtered_purchases_sample = np.std(filtered_purchases, ddof = 1)
print(std_filtered_purchases_sample)
```

```python
mean_of_filtered_visits = np.mean(filtered_visits)
print(mean_of_filtered_visits)
std_filtered_visits_sample = np.std(filtered_visits, ddof = 1)
print(std_filtered_visits_sample)
```

Коваріація filtered purchases and filtered visits

```python
filtered_purchases_visits = [(filtered_purchases[i] - mean_of_filtered_visits) * (filtered_visits[i] - mean_of_filtered_visits) for i in range (len(filtered_purchases))]
cov_filtered_purchases_visits = sum(filtered_purchases_visits)/(len(filtered_purchases) - 1)
```

```python
cov_filtered_purchases_visits
```

```python
np.var(filtered_purchases, ddof = 1)
```

```python
np.cov(filtered_purchases, filtered_visits)
```

Розрахунок кореляції Пірсона між filtered puchases and filtered visits

```python
cov_filtered_purchases_visits / (std_filtered_purchases_sample * std_filtered_visits_sample)
```

```python
np.corrcoef(filtered_purchases, filtered_visits)
```
