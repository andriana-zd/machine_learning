1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 10 * x - 5
```

```python
def f_derivative(x):
    return np.full_like(x, 10)
```

```python
x = np.linspace(0, 10, 100)
```

```python
y = f(x)
y_derivative = f_derivative(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = 10x - 5', color='blue', linewidth=2)  # Графік функції
plt.plot(x, y_derivative, label="f'(x) = 10", color='red', linestyle='--', linewidth=2)  # Графік похідної

plt.title('Графік функції та її похідної', fontsize=16)
plt.xlabel('Кількість годин навчання (x)', fontsize=14)
plt.ylabel('Оцінка', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Горизонтальна лінія
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Вертикальна лінія
plt.ylim(-20, 120)
plt.xlim(0, 10)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, ln, diff, lambdify
```

```python
x = symbols('x')
f1 = sin(x)

f1_derivative = diff(f1, x)

f1_func = lambdify(x, f1, modules=['numpy'])
f1_derivative_func = lambdify(x, f1_derivative, modules=['numpy'])

x_vals_f1 = np.linspace(-2 * np.pi, 2 * np.pi, 500)

y_vals_f1 = f1_func(x_vals_f1)
y_vals_f1_derivative = f1_derivative_func(x_vals_f1)

plt.figure(figsize=(6, 6))
plt.plot(x_vals_f1, y_vals_f1, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals_f1, y_vals_f1_derivative, label="f'(x) = cos(x)", color='red', linestyle='--')
plt.title('Функція sin(x) та її похідна', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

```python
f2 = ln(x)

f2_derivative = diff(f2, x)

f2_func = lambdify(x, f2, modules=['numpy'])
f2_derivative_func = lambdify(x, f2_derivative, modules=['numpy'])

x_vals_f2 = np.linspace(0.1, 10, 500)

y_vals_f2 = f2_func(x_vals_f2)
y_vals_f2_derivative = f2_derivative_func(x_vals_f2)

plt.figure(figsize=(6, 6))
plt.plot(x_vals_f2, y_vals_f2, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals_f2, y_vals_f2_derivative, label="f'(x) = 1/x", color='red', linestyle='--')
plt.title('Функція ln(x) та її похідна', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, lambdify
```

```python
x = symbols('x')
f = x**2 - 5*x + 1
```

```python
f_derivative = diff(f, x)
```

```python
f_func = lambdify(x, f, modules=['numpy'])
f_derivative_func = lambdify(x, f_derivative, modules=['numpy'])
```

```python
x_vals = np.linspace(0, 100, 500)
```

```python
y_vals = f_func(x_vals)
y_derivative_vals = f_derivative_func(x_vals)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x) = x^2 - 5x + 1', color='blue', linewidth=2)
plt.plot(x_vals, y_derivative_vals, label="f'(x) = 2x - 5", color='red', linestyle='--', linewidth=2)

plt.title('Функція витрат та її похідна', fontsize=16)
plt.xlabel('Кількість вироблених одиниць (x)', fontsize=14)
plt.ylabel('Витрати', fontsize=14)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
