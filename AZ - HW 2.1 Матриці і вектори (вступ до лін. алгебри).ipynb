### –ó–∞–≤–¥–∞–Ω–Ω—è 1: –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—å —Ç–∞ –≤–µ–∫—Ç–æ—Ä—ñ–≤
1. –°—Ç–≤–æ—Ä—ñ—Ç—å `numpy` –≤–µ–∫—Ç–æ—Ä `vector` —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—é 10 –∑ –≤–∏–ø–∞–¥–∫–æ–≤–∏–º–∏ –¥—ñ–π—Å–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏ –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –≤—ñ–¥ 0 –¥–æ 1. –ú–æ–∂–Ω–∞ —Å–∫–æ—Ä–∏—Å—Ç–∞—Ç–∏—Å—å –¥–≤–æ–º–∞ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∏–º–∏ –≤–∞—Ä—ñ–∞–Ω—Ç–∞–º–∏:
  1. –ì–µ–Ω–µ—Ä–∞—Ü—ñ—è –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö —á–∏—Å–µ–ª –∑ –º–æ–¥—É–ª–µ–º [`random.uniform`](https://docs.python.org/3/library/random.html).
  2. –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä–∞ –∑ –¥–æ–ø–æ–º–æ–≥–æ—é –º–µ—Ç–æ–¥–∞ `numpy` [np.random.rand](https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html)

2. –°—Ç–≤–æ—Ä—ñ—Ç—å –º–∞—Ç—Ä–∏—Ü—é `matrix` —Ç–∏–ø—É `numpy.array` —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—ñ 3x3 –∑ –Ω–∞—Å—Ç—É–ø–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏

  $$
  matrix = \begin{bmatrix}
  11 & 14 & 17 \\
  20 & 23 & 26 \\
  29 & 32 & 35
  \end{bmatrix}
  $$

  –±—É–¥—å-—è–∫–∏–º –≤—ñ–¥–æ–º–∏–º–∏ –í–∞–º —Å–ø–æ—Å–æ–±–æ–º.

–í–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –≤–µ–∫—Ç–æ—Ä —ñ –º–∞—Ç—Ä–∏—Ü—é, –∞ —Ç–∞–∫–æ–∂ —ó—Ö —Ä–æ–∑–º—ñ—Ä–∏.


```python
import numpy as np
```

```python
vector = np.random.rand(10)
```

```python
matrix = np.array([
    [11, 14, 17],
    [20, 23, 26],
    [29, 32, 35]
])
```

```python
print("–í–µ–∫—Ç–æ—Ä:")
print(vector)
print("–†–æ–∑–º—ñ—Ä –≤–µ–∫—Ç–æ—Ä–∞:", vector.shape)
```

```python
print("\n–ú–∞—Ç—Ä–∏—Ü—è:")
print(matrix)
print("–†–æ–∑–º—ñ—Ä –º–∞—Ç—Ä–∏—Ü—ñ:", matrix.shape)
```


### –ó–∞–≤–¥–∞–Ω–Ω—è 2: –û–ø–µ—Ä–∞—Ü—ñ—ó –Ω–∞–¥ –≤–µ–∫—Ç–æ—Ä–∞–º–∏ —Ç–∞ –º–∞—Ç—Ä–∏—Ü—è–º–∏
–í–∏–∫–æ–Ω–∞–π—Ç–µ –∫–æ–∂–Ω—É –∑ –∑–∞–∑–Ω–∞—á–µ–Ω–∏—Ö –æ–ø–µ—Ä–∞—Ü—ñ–π –Ω–∞–¥ –∑–∞–∑–Ω–∞—á–µ–Ω–∏–º–∏ –æ º—î–∫—Ç–∞–º–∏ —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω:

3. –°—Ç–≤–æ—Ä—ñ—Ç—å –Ω–æ–≤–∏–π –≤–µ–∫—Ç–æ—Ä, —è–∫–∏–π –º—ñ—Å—Ç–∏—Ç—å –∫–≤–∞–¥—Ä–∞—Ç –∫–æ–∂–Ω–æ–≥–æ `vector` –±–µ–∑ —Ü–∏–∫–ª—ñ–≤.


### –ó–∞–≤–¥–∞–Ω–Ω—è 3: –ú–Ω–æ–∂–µ–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—å
1. –ó–∞–¥–∞–Ω—ñ –¥–≤—ñ –º–∞—Ç—Ä–∏—Ü—ñ `A` —ñ `B`—Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—è–º–∏ 3x2 —Ç–∞ 2—Ö3. –í–∏–∫–æ–Ω–∞–π—Ç–µ —ó—Ö–Ω—î –º–Ω–æ–∂–µ–Ω–Ω—è, —â–æ–± –æ—Ç—Ä–∏–º–∞—Ç–∏ –º–∞—Ç—Ä–∏—Ü—é 2x2. –í–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç—É—é—á—É –º–∞—Ç—Ä–∏—Ü—é `result_matrix` –Ω–∞ –µ–∫—Ä–∞–Ω.

2. –ó–∞–¥–∞–Ω–∏–π –≤–µ–∫—Ç–æ—Ä `C` —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—é 2—Ö1. –ü–æ–¥—É–º–∞–π—Ç–µ —Å–ø–æ—á–∞—Ç–∫—É, —è–∫—É —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –í–∏ –æ—Ç—Ä–º–∞—î—Ç–µ, –∫–æ–ª–∏ –¥–æ–º–Ω–æ–∂–∏—Ç–µ `result_matrix` –Ω–∞ –≤–µ–∫—Ç–æ—Ä `C` —ñ —è–∫–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ. –ê –ø–æ—Ç—ñ–º –≤–∏–∫–æ–Ω–∞–π—Ç–µ –º–Ω–æ–∂–µ–Ω–Ω—è —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω.

3. –ó–∞–¥–∞–Ω–∞ –º–∞—Ç—Ä–∏—Ü—è `D` —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—é 2—Ö2. –ü–æ–¥—É–º–∞–π—Ç–µ —Å–ø–æ—á–∞—Ç–∫—É, —è–∫—É —Ä–æ–∑–º—ñ—Ä–Ω—ñ—Å—Ç—å –í–∏ –æ—Ç—Ä–º–∞—î—Ç–µ, –∫–æ–ª–∏ –¥–æ–º–Ω–æ–∂–∏—Ç–µ `result_matrix` –Ω–∞ –º–∞—Ç—Ä–∏—Ü—é `D` —ñ —è–∫–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ. –ê –ø–æ—Ç—ñ–º –≤–∏–∫–æ–Ω–∞–π—Ç–µ –º–Ω–æ–∂–µ–Ω–Ω—è —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω.

```python
A = np.array(
    [[0.18, 0.53],
     [0.75, 0.47],
     [0.35, 0.21]])

B = np.array(
    [[0.17, 0.58, 0.75],
     [0.38, 0.11, 0.15]])

C = np.array([[0], [1]])

D = np.array([[0,1], [1,0]])

display(A, B, C, D)
```

```python
import numpy as np
```

```python
result_matrix = np.dot(A, B)
print(result_matrix)
```

–ú–Ω–æ–∂–µ–Ω–Ω—è result_matrix –Ω–∞ –≤–µ–∫—Ç–æ—Ä C


```python
result_matrix_sliced = result_matrix[:, :2]

final_result_vector = np.dot(result_matrix_sliced, C)
print(final_result_vector)
```

–ú–Ω–æ–∂–µ–Ω–Ω—è result_matrix –Ω–∞ –º–∞—Ç—Ä–∏—Ü—é ùê∑


```python
final_result_matrix = np.dot(result_matrix_sliced, D)
print(final_result_matrix)
```


### –ó–∞–≤–¥–∞–Ω–Ω—è 4: –¢—Ä–∞–Ω—Å–ø–æ–Ω—É–≤–∞–Ω–Ω—è –º–∞—Ç—Ä–∏—Ü—å
1. –¢—Ä–∞–Ω—Å–ø–æ–Ω—É–π—Ç–µ `result_matrix`.
2. –ó—Ä–æ–±—ñ—Ç—å —Ç—Ä–∞–Ω—Å–ø–æ–Ω—É–≤–∞–Ω–Ω—è –≤–µ–∫—Ç–æ—Ä–∞ `vector` —ñ –≤–∏–≤–µ–¥—ñ—Ç—å —Ñ–æ—Ä–º—É –Ω–æ–≤–æ—ó —Å—Ç—Ä—É–∫—Ç—É—Ä–∏. –ß–∏ –±–∞—á–∏—Ç–µ –≤–∏ –∑–º—ñ–Ω—É?
3. –í–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ –æ–ø–µ—Ä–∞—Ü—ñ—é `numpy.expand_dims` –Ω–∞–¥ –≤–µ–∫—Ç–æ—Ä–æ–º `vector` –∑ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º axis=1. –ó–∞–ø–∏—à—ñ—Ç—å –æ—Ç—Ä–∏–º–∞–Ω–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –∑–º—ñ–Ω–Ω—É `column_vector` —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω. –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ –∑ `vector`.
4. –ó—Ä–æ–±—ñ—Ç—å —Ç—Ä–∞–Ω—Å–ø–æ–Ω—É–≤–∞–Ω–Ω—è `column_vector`, –≤–∏–≤–µ–¥—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ –µ–∫—Ä–∞–Ω —Ä–∞–∑–æ–º –∑ —Ä–æ–∑–º—ñ—Ä–Ω–æ—Å—Ç—è–º–∏ –æ—Ç—Ä–∏–º–∞–Ω–æ—ó —Å—Ç—Ä—É–∫—Ç—É—Ä–∏.




```python
transposed_matrix = result_matrix.T
print(transposed_matrix)
```

```python
transposed_vector = vector.T

print(transposed_vector)
print("–§–æ—Ä–º–∞ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–æ–≤–∞–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞:", transposed_vector.shape)
print("–§–æ—Ä–º–∞ –æ—Ä–∏–≥—ñ–Ω–∞–ª—å–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ vector:", vector.shape)
```

–£ –æ–¥–Ω–æ–≤–∏–º—ñ—Ä–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞ —Ç—Ä–∞–Ω—Å–ø–æ–Ω—É–≤–∞–Ω–Ω—è –Ω–µ –∑–º—ñ–Ω–∏–ª–æ —Ñ–æ—Ä–º—É.

```python
column_vector = np.expand_dims(vector, axis=1)

print(column_vector)
print("–§–æ—Ä–º–∞ column_vector:", column_vector.shape)

print("–û—Ä–∏–≥—ñ–Ω–∞–ª—å–Ω–∏–π vector:")
print(vector)
print("–§–æ—Ä–º–∞ vector:", vector.shape)
```

–í—ñ–¥–±—É–ª–æ—Å—å –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —É —Å—Ç–æ–≤–ø—Ü–µ–≤–∏–π –≤–µ–∫—Ç–æ—Ä

```python
transposed_column_vector = column_vector.T

print("–¢—Ä–∞–Ω—Å–ø–æ–Ω–æ–≤–∞–Ω–∏–π column_vector:")
print(transposed_column_vector)
print("–§–æ—Ä–º–∞ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–æ–≤–∞–Ω–æ–≥–æ column_vector:", transposed_column_vector.shape)
```
